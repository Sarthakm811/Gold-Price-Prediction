```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

```python
gold_data=pd.read_csv("/content/gld_price_data.csv")
```

```python
gold_data.head()
```

```python
gold_data.describe()
```

```python
gold_data.drop(columns="Date", inplace=True)
```

```python
correlation=gold_data.corr()
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap="Blues")
```

```python
correlation["GLD"]
```

```python
#checking the distribution of the GLD price
sns.distplot(gold_data["GLD"],color="green")
```

Splitting the features and target

```python
x= gold_data.drop(['GLD'],axis=1)
y= gold_data["GLD"]
```

```python
x
```

```python
y
```

Splitting into Training data and Test data

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

Model Training

```python
#Random forest
```

```python
regressor=RandomForestRegressor(n_estimators=100)
```

```python
regressor.fit(X_train,Y_train)
```

Model evalution

```python
test_data_prediction=regressor.predict(X_test)
```

```python
test_data_prediction.mean()
```

```python
#R squre erorr
error_score=metrics.r2_score(Y_test,test_data_prediction)
```

```python
error_score
```

Compare the actual value and Predicted value

```python
Y_test=list(Y_test)
```

```python
plt.plot(Y_test,color="k",label="Actual Value")
plt.plot(test_data_prediction,color="y",label="Predicted Value")
plt.xlabel("Number of Values")
plt.ylabel("Gold Price")
plt.legend()
plt.show()
```
